```python
xmin = 133.8882
ymin = 92.5204
xmax = 342.4848
ymax = 197.8627
# img = cv2.resize(img,(img.shape[0],img.shape[1]))
img_crop = img_cvt[int(np.around(ymin))-6:int(np.around(ymin))+int(np.around(ymax-ymin))+7,int(np.around(xmin)):int(np.around(xmin))+int(np.around(xmax-xmin))]
plt.imshow(img_crop)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
img_crop_gray = cv2.cvtColor(img_crop,cv2.COLOR_BGR2GRAY)
# for the surface map
ax = plt.axes(projection='3d')
y = range(img_crop.shape[0] )
x = range( img_crop.shape[1] ) 
X, Y = np.meshgrid(x, y)
ax.plot_surface( X, Y, img_crop_gray,cmap=cm.coolwarm) # Z axis has the 
ax.view_init(90,90)
plt.show()
# plt.savefig("NYU0015_disp_plot.png")
```

```python
# img_crop_gray = cv2.cvtColor(img_crop,cv2.COLOR_BGR2GRAY)
# for the surface map
img_crop_gray_new = img_crop_gray
# img_crop_gray_new[img_crop_gray_new>140] = 255
result = np.clip(img_crop_gray,60,170)
print(img_crop_gray_new[0])
ax = plt.axes(projection='3d')
y = range(img_crop_gray_new.shape[0] )
x = range(img_crop_gray_new.shape[1] ) 
X, Y = np.meshgrid(x, y)
ax.plot_surface( X, Y, result,cmap=cm.coolwarm) # Z axis has the 
ax.view_init(90,90)
plt.show()
```
