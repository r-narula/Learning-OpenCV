```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data (1).csv")
```

```python
print(data)
```

```python
days=np.array(data["day"])
print(days.T.shape)
```

```python
deaths = np.array(data[" deaths"])
new_deaths=deaths.reshape(deaths.size,1)
new_days=days.reshape(days.size,1)
x0=np.ones((days.size,1))
Days=np.concatenate(( x0,new_days), axis=1)
```

```python
theta=np.dot(np.dot(np.linalg.inv(np.dot(Days.T,Days)), Days.T) , new_deaths)
print(theta)
```

```python
def abline(slope, intercept):
    print(slope,intercept)
    x = np.linspace(0,300,300)
    y=slope*x+intercept
    plt.plot(x, y,'-r')
    plt.plot(days, deaths, color= "green") 
    
    
    # function to show the plot 
    plt.show() 
```
