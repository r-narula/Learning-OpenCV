Calculating the gradient intensity 
Image gradient is the directional change in the intensity or color in an image.
measures how the image is changing.
Magnitude of gradient tells how quickly the image is changing, while direction tells in which direction the image is changing most rapidly.


```python
img = cv2.imread("flower.jpg")
imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
laplacian = cv2.Laplacian(imgGray,cv2.CV_8U)
sobelx = cv2.Sobel(imgGray,cv2.CV_8U,ksize=5,dx=1,dy=0)
sobely = cv2.Sobel(imgGray,cv2.CV_8U,0,1,ksize=5)
# Doubt
# I am  not able to understand the use fo cv2.CV_8U,cv2CV_64F
```

## Steps for Canny edge detection
1) Noise reduction -> Done by smoothning an image.

2) Intensity Gradient -> image is made to pass through Sobel kernel in        both horizontal and vertical direction.

3) Now we supress the areas where the edge (perpendicular) value is less than the gradient (parallel).

4) Then the thresholding is Done so as to make more edges or less . 


```python
imgCanny = cv2.Canny(img,50,150)
plt.imshow(imgCanny,cmap="gray")
```

## Dilation

Adds pixels to the boundaries of an image and the erosion removes the pixels from the boundaries. 

increase the thickness of the edge.



```python
j = cv2.imread("j.png")
grayj = cv2.cvtColor(j,cv2.COLOR_BGR2GRAY)
plt.imshow(grayj,cmap="gray")
```

```python
dilimg = cv2.dilate(grayj,kernel=np.ones((5,5),dtype=np.int32),iterations=1)
plt.imshow(dilimg,cmap="gray")
```

```python
## eroding an image
imgErode = cv2.erode(grayj,kernel=np.ones((5,5),dtype=np.int32),iterations=1)
plt.imshow(imgErode,cmap="gray")
```

Resizing and the scaling are different 

In Resizing we change the pixel information .
While increasing the size two methods are common ->

1)Nearest Neighbour - In nearest neightbour we make the new pixel equal to the nearest neighbour and thus this leads to pixelation in the image.

2) Bilinear and bicubic -> In the range of 2*2 it increases the pixel by taking the weighted average of all the 4 pixels and then marks it there.

Bicubic Interpolation makes by taking 4*4 pixels (16 in total) and then taking weighted mean by givin the closer ones more weight ....

## Putting text on Images

```python
img = cv2.imread("Lenna.png")
```

```python
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 
  
# org 
org = (50, 50) 
  
# fontScale 
fontScale = 1
   
# Blue color in BGR 
color = (255, 0, 0) 
  
# Line thickness of 2 px 
thickness = 2
   
# Using cv2.putText() method 
image = cv2.putText(img, 'OpenCV', org, font,fontScale, color, thickness, cv2.LINE_AA) 
```

```python
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
cv2.imshow("image",image)
cv2.waitKey(0) 
cv2.destoryAllWindows() 
```
