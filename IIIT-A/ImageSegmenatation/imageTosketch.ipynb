```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("House.jpeg")
```

```python
plt.imshow(img)
```

```python
# blur_img = cv2.GaussianBlur(img,(3,3),0)
#plt.imshow(blur_img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,"gray")
```

```python
edge_img = cv2.Canny(img,67,100)
plt.imshow(edge_img,"gray")
```

```python
# ## thresholding adaptive 
# th3 = cv2.adaptiveThreshold(edge_img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
# plt.imshow(th3,"gray")
```

```python
new = 255 - edge_img
```

```python
final = cv2.GaussianBlur(new,(5,5),0)
```

```python
plt.imshow(final,"gray")
```

```python
# using to calculate the gradient intensity of the image.
laplacian = cv2.Laplacian(gray,cv2.CV_8U)
plt.imshow(laplacian,"gray")
```

```python
laplacian_invert = 255 - laplacian
```

```python
plt.imshow(laplacian_invert,"gray")
```
