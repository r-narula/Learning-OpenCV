```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("newRaghav.jpeg")
```

```python
plt.imshow(img)
```

```python
blur_img = cv2.GaussianBlur(img,(7,7),0)
```

```python
plt.imshow(blur_img)
```

```python
gray = cv2.cvtColor(blur_img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,"gray")
```

```python
edge_img = cv2.Canny(img,50,100)
plt.imshow(edge_img,"gray")
```

```python
## thresholding adaptive 
th3 = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
plt.imshow(th3,"gray")
```

```python
new = 255 - edge_img
```

```python
final = cv2.GaussianBlur(new,(7,7),0)
```

```python
plt.imshow(final,"gray")
```
