```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Raghav.jpeg")
```

```python
plt.imshow(img)
```

```python
blur_img = cv2.GaussianBlur(img,(7,7),0)
```

```python
plt.imshow(blur_img)
```

```python
gray = cv2.cvtColor(blur_img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,"gray")
```

```python
edge_img = cv2.Canny(img,40,60)
plt.imshow(edge_img,"gray")
```

```python
new = 255 - edge_img
```

```python
final = cv2.GaussianBlur(new,(5,5),0)
```

```python
plt.imshow(final,"gray")
```

```python
plt.imsave("raghavFinal.jpg",final)
```
